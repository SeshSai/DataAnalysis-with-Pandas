```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.options.display.max_columns = 500
pd.options.display.max_rows = 500
pd.options.display.float_format = '{:.6f}'.format
```

```python
data = pd.read_csv("movies_complete.csv")
```

```python
#Ahh the traditions, lets view the data frame. Just five rows to get a feel for it
data.head(5)
```

#### As a film journalists, we are interested in finding how Franchise movies and stand alone movies perform

#### Few questions we can ask are:
    Which Movie franchise is most profitable
    Franchise vs non franchise movies avg budget
    Franchise vs non franchise movies avg revenue
    Franchise vs non franchise movies ROI
    Franchise vs non franchise movies Average Rating

```python
#,notna() method returns booleans. If na False, If not null True
data['franchise'] = data["belongs_to_collection"].notna()
```

```python
data['franchise'].value_counts()
```

```python

data_franchise = data.loc[data['franchise'] == True]
```

```python
data_franchise['Roi'] = data_franchise.revenue_musd.div(data_franchise.budget_musd)
```

```python
#Franchise vs single movies avg revenue
data.groupby('franchise').budget_musd.mean()
```

```python
#Franchise vs single movies avg revenue
data.groupby('franchise').revenue_musd.mean()
```

```python
data['Roi'] = data.revenue_musd.div(data.budget_musd)
```

```python
#Franchise vs non franchise movies ROI
data.groupby('franchise').Roi.median()
```

```python
#Franchise vs non franchise movies Average Rating
data.groupby('franchise').vote_average.mean()
```

```python
#Now let us look at all metrics together and draw some conclusion and we will use filter to select only movies with atleast
#50 votes

data.groupby(data['franchise'].loc[data['vote_count'] >= 50]).agg({'budget_musd':'mean','revenue_musd':'mean','Roi':'median','vote_average':'mean','vote_count':'mean'})
```

#### Now let us see which franchise is the most successful in terms of ROI and rating

```python
franchise_data = data.loc[data['belongs_to_collection'].notna()].reset_index(drop = True)
```

```python
franchise_data['Roi'] = franchise_data.revenue_musd.div(franchise_data.budget_musd)
```

```python
franchise_data.groupby(franchise_data['belongs_to_collection'].loc[(franchise_data['budget_musd']>=5) & 
                                                                   (franchise_data['Roi'].notna())]).agg({'Roi':'mean'}).sort_values(by = 'Roi',ascending = False).reset_index()
```

```python
franchise_data.groupby(franchise_data['belongs_to_collection'].loc[(franchise_data['vote_count']>=100) & 
                                                                   (franchise_data['vote_average'].notna())]).agg({'vote_average':'mean'}).sort_values(by = 'vote_average',ascending = False).reset_index(drop = True)
```

#### We can further filter us by language, country etc. But we got the idea on how basic pandas is useful to extract interesting insights from data

#### Now let us extract the same info using multi index which makes our live a bit easier

```python
franchise_data.groupby(franchise_data['belongs_to_collection'].loc[(franchise_data['vote_count']>=100) & 
                                                                   (franchise_data['vote_average'].notna())]).agg({'vote_average':'mean'}).sort_values(by = 'vote_average',ascending = False).reset_index(drop = True)
```

```python
frachise_data = data.groupby(data['belongs_to_collection'].loc[(data['belongs_to_collection'].notna()) & 
                                                               (data['budget_musd']>=5)]).agg({'title':'count',
                                                                                                           'budget_musd':['sum','mean'],
                                                                                                           'revenue_musd':['sum','mean'],
                                                                                                            'Roi':'median',
                                                                                                           'vote_average':'mean'})
```

```python
frachise_data
```

#### Now Your boss wants to find out: 
    which franchise has the highest ROI
    Which franchise has 5 movies or more presented by lowest to highest ROI
    Which franchise has highest rating
    Which franchise has rating > 6 or more presented by lowest to highest ROI

```python
frachise_data.nlargest(10,('Roi', 'median'))
```

```python
frachise_data[frachise_data[('title','count')]>= 5].sort_values(by=('Roi', 'median'),ascending = True).head(5)
```

```python
frachise_data.nlargest(10,('vote_average', 'mean')).head(5)
```

#### nlargest works similar to sort_values method. We will use that instead of sort_values below

```python
frachise_data[frachise_data[('vote_average','mean')]>=6].nlargest(10,('Roi', 'median'))
```
