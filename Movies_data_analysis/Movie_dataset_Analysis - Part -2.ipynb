```python
#,notna() method returns booleans. If na False, If not null True
data['franchise'] = data["belongs_to_collection"].notna()
```

```python
#Franchise vs non franchise movies Average Rating
data.groupby('franchise').vote_average.mean()
```

```python
#Now let us look at all metrics together and draw some conclusion and we will use filter to select only movies with atleast
#50 votes

data.groupby(data['franchise'].loc[data['vote_count'] >= 50]).agg({'budget_musd':'mean','revenue_musd':'mean','Roi':'median','vote_average':'mean','vote_count':'mean'})
```

```python
franchise_data = data.loc[data['belongs_to_collection'].notna()].reset_index(drop = True)
```

#### There are few challenges here in processing our data
    Cast column has all the actors for each row i.e is each movie seperated by a pipe("|").
    We need to seperate each actor into new columns and plan our approach on how to deal with this

```python
data.cast.head(5)
```

```python
data.cast.str.split("|").head(5)
```

#### Now I want to separate these seperated actor names into new columns
#### Pandas has expand parameter in slpit method to help with this

```python
data.cast.str.split("|",expand = True)
```

#### Oh, damn! One movie had more than 300 main actors or may be total cast??

#### Okay, now how are we going to approach this to answer our questions??

#### One way is we will transpose this in a way that we get actor names as columns and every row is filled with binary value (1,0). This indicates wether the actor is present that particular movie or not

#### Now the problem is if we have tom hanks as a second actor in one movie and 10th actor in another movie??

```python
actors = data.cast.str.split("|",expand = True)
```

```python
####How about having movie id as index which is unique
```

```python
data.set_index("id",inplace = True)
```

```python
data.head(5)
```

```python
actors = data.cast.str.split("|",expand = True)
```

```python
actors.head(5)
```

```python
#### Okay now we have id as index and each id has actors in seperate columns
#### We will use stack method which converts the columns into index with the existing index as multilevel
```

```python
actors.stack()
```

```python
#Now the inner level 1 index is not so useful which is repeating for every outer level index
#We will drop that column
actors = actors.stack().reset_index(level = 1, drop = True).to_frame()
```

```python
### Now we will rename our column to Actor
actors.columns = ['Actor']
```

```python
actors.head(5)
```

```python
#Now based on film id index we will take required columns from original data frame and merge them.
#The merge is done in way that the resultant data frame will have all the rows from the actors data frame and we will add other data from
#the original data frame where the film id matches.

actors = actors.merge(data[['title','revenue_musd','Roi','vote_count','vote_average','popularity']],how = 'left',left_index = True,right_index = True)
```

```python
actors.head(5)
```

#### Which actor acted in highest number of movies
#### Which actor had highest revenue

```python

actors_1 = actors.groupby('Actor').agg(Total_Revenue =("revenue_musd","sum"),
                            Total_movies = ("Actor","count")).reset_index()
                            
```

```python
actors_1[actors_1['Total_movies'] > 10].sort_values('Total_Revenue',ascending = False).head(10)
```
